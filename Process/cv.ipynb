```python
import requests
import pandas as pd

urls = pd.read_csv('../data/updated_products.csv')['image']
```

```python
valid_urls = []
def valid_url(url):
    try:
      response = requests.get(url)
      response.raise_for_status()
      valid_urls.push(url)
      return True
    except Exception as e:
       return None 
urls.apply(lambda x: 1 if valid_url(x) else 0)
```
